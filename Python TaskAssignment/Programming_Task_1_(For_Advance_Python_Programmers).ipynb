# Q1 Sum Of eliments of list

```python
class Solution:
    def runningSum(self, nums):
        for i in range(1, len(nums)):
            nums[i] += nums[i - 1]
        return nums
s = Solution()
s.runningSum([1,2,3,4])
```

# Q2 Resuffle list

```python
'''
n = total elements of both side x,y
suppose we have list [1,2,3,4,5,6] so n = 3
i+n means = i = 1, n= 3 then we have i+n = 4
first i = 1 then i+n =4
[1,4]
[1,4,2,5]...so on
'''

class Solution:
    def shuffle(self, nums, n):
        flist = []

        for i in range(n):
            flist.append(nums[i])
            flist.append(nums[i+n])
        return flist

a = Solution()
a.shuffle([1,2,3,4,5,6],3)
```

# Q3 Extra Candies

```python
class Solution:
    def kidsWithCandies(self, candies, extraCandies):
         m = max(candies)
         result = []
         for i in candies:
            if (i+extraCandies >= m):
                result.append(True)
            else:
                result.append(False)
         return result


c = Solution()
c.kidsWithCandies([2,3,5,6,1,1,2,4,3],3)
```
